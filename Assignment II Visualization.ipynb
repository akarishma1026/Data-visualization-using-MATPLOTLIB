# VISUALIZATION USING SEABORN AND MATPLOTLIB

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sale=pd.read_csv("D:/II Semester/Python/Supermarket Grocery Sales (Karishma).csv")
```

```python
#dataset of grocery sales
sale
```

```python
x=list(sale.iloc[:,6])
y=list(sale.iloc[:,7])
plt.bar(x,y,color='aquamarine')
plt.title("Region_Vs_Sales")
plt.xlabel("Region")
plt.ylabel("Sales")
plt.xticks(color="magenta",rotation=90,fontsize='12',verticalalignment='top')

plt.show()
```

### Here we are comparing the sales region wise through which we can find that South, West, Central, East are almost have the same amount of sales whereas the the North region saless in need to be improved

```python
plt.boxplot(sale['Profit'])
```

### In this boxplot we have taken the variable sales. Through this we can understand that the minimum sales is approximately Rs.30 whereas the maximum is above Rs.1000. The average sales of the grocery market is Rs. 300. Most of the sales are in good amount

```python
plt.hist(sale['Discount'],color='thistle')
plt.show()
```

### Through thsi histogram we get to know that the discount range is between 0.10 to 0.35. Here maximum sales have got the discount between 0.25 to 0.27 and 0.37 to 0.35. 

```python
plt.scatter(sale['Sales'],sale['Profit'],color='midnightblue')
plt.xlabel('Profit')
plt.ylabel('Sale')
plt.xticks(color='gold',fontsize='12',fontweight='bold')

plt.show()
```

### This scatter plot shows the relationship between the sales and the profit. Observing the plot we can assure that they both have positive relationship the higher the sale the more the profit.

```python
sns.histplot(sale['Sub Category'],color='beige')
plt.xlabel('Sub Category')
plt.xticks(color='g',fontsize='12',fontweight='bold',rotation="90")

plt.show()
```

### This histogram tells about the sale according to sub category. The most bought product is under the category of Masala and soft drinks. The least bought category is rice. All the other categories have average or equal number of sales

```python
sns.distplot(sale['Sales'],color='hotpink')
plt.show()
```

### Looking into the density plot we can find that the most of the sales are densed between 500 to 2500. Most of the sales happen within these range

```python
sns.distplot(sale['Sales'][:300],rug= True, color='grey')
plt.show
```

### In this distplot we look into the first 300 obervations of sales and we get to know that the again the maximum sales are between 500 and 2500 . If we look closer into the density line the major sales are around 700 to 1500.

```python
sns.distplot(sale['Profit'],hist=False)
```

### This distplot shows us that the most profit that the supermarket gets is between 100 to 400. We can observe that they have faced little loss too though has good stable profit

```python
#subplot1

plt.subplot(2,2,1)
plt.title('Sales')
sns.distplot(sale['Sales'],color='violet')

#subplot2

plt.subplot(2,2,2)
plt.title('Discount')
sns.distplot(sale['Discount'],color='hotpink')


#Subplot 3

plt.subplot(2,2,3)
plt.title('Profit')
sns.distplot(sale['Profit'],color='thistle')

plt.tight_layout()
```

### Subplots are used to compare the data in a single sight. The maximum sales are densed around 1000 to 2000. the discounts the customers get are between 0.10 to 0.25. The most profit the grocery market gains are around 400.

```python
sns.boxplot(y=sale['Discount'],color='cyan')
plt.title('Discount')

plt.show()
```

### Through this boxplot we can find that the minimum discount is 0.10 and the maximum is 0.35. The average discount the supermarket offers is 0.23.

```python
#joint plots

sns.jointplot('Sales','Profit',sale,color='gold',kind='kde')
plt.show()
```

```python
sns.jointplot('Sales','Profit',sale,color='darkslateblue',kind='scatter')
```

```python
sns.jointplot('Sales','Profit',sale,color='magenta',kind='reg')
```

```python
sns.jointplot('Sales','Profit',sale,color='indigo',kind='hex')
```

### All these map are the varieties of joint plot. Here we compare the sales and profit and we get to know that the both have the positive relationship. Heat map and hex shows us the range between 500 to 1000 are densed comparatively to other maps
